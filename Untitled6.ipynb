```python

# ======================================================
# Secure Multi-Party Computation (SMPC) - Lab Assignment
# Special Topics in Information Security
# Seiyun University, Yemen
# Instructor: Prof. Ahmed Abuamer
# ======================================================

#@title **ุงูุฌุฒุก 0: ุงุณุชูุฑุงุฏ ุงูููุชุจุงุช ุงููุทููุจุฉ**
import random
from typing import List, Tuple, Dict

print("=" * 70)
print("      Secure Multi-Party Computation (SMPC) - Lab Assignment")
print("=" * 70)
print()

#@title **ุงูุฌุฒุก 1: ุญุณุงุจ ุงููุฌููุน ุจุงุณุชุฎุฏุงู ุชูุงุณู ุงูุณุฑ (MPC Sum)**
def generate_random_shares(value: int, num_parties: int) -> List[int]:
    """ุชูุณู ุงููููุฉ ุงูุณุฑูุฉ ุฅูู ุญุตุต ุนุดูุงุฆูุฉ."""
    shares = []
    sum_of_shares = 0

    for i in range(num_parties - 1):
        share = random.randint(0, 100)
        shares.append(share)
        sum_of_shares += share

    last_share = value - sum_of_shares
    shares.append(last_share)

    return shares

def mpc_sum_experiment():
    """ุชุฌุฑุจุฉ ุญุณุงุจ ูุฌููุน ุฑูุงุชุจ ุซูุงุซุฉ ุฃุดุฎุงุต ุจุงุณุชุฎุฏุงู MPC."""
    print("[1] ุชุฌุฑุจุฉ ุญุณุงุจ ุงููุฌููุน ุจุงุณุชุฎุฏุงู ุชูุงุณู ุงูุณุฑ (MPC Sum):")
    print("-" * 60)

    private_values = {
        "Ahmed": 75,
        "Ali": 85,
        "Rashida": 60
    }

    print("ุงูููู ุงูุณุฑูุฉ ููู ูุดุงุฑู:")
    for person, salary in private_values.items():
        print(f"  {person}: {salary}")
    print()

    # ุชูููุฏ ุงูุญุตุต
    shares = {}
    for person, value in private_values.items():
        shares[person] = generate_random_shares(value, len(private_values))

    print("ุงูุญุตุต ุงูุนุดูุงุฆูุฉ ุงูุชู ุชููุฏูุง ูู ูุดุงุฑู:")
    for person, share_list in shares.items():
        print(f"  {person}: {share_list}")
    print()

    # ุชูุฒูุน ุงูุญุตุต
    received_shares = {person: [] for person in private_values.keys()}
    parties = list(private_values.keys())

    for i, giver in enumerate(parties):
        for j, receiver in enumerate(parties):
            if i == j:
                continue
            received_shares[receiver].append(shares[giver][j])

    print("ุงูุญุตุต ุงูุชู ูุณุชูููุง ูู ูุดุงุฑู:")
    for person, share_list in received_shares.items():
        print(f"  {person}: {share_list}")
    print()

    # ุญุณุงุจ ุงููุฌุงููุน ุงูุฌุฒุฆูุฉ
    partial_sums = {}
    for person, share_list in received_shares.items():
        partial_sums[person] = sum(share_list)

    print("ุงููุฌุงููุน ุงูุฌุฒุฆูุฉ ุงูุชู ูุญุณุจูุง ูู ูุดุงุฑู:")
    for person, partial_sum in partial_sums.items():
        print(f"  {person}: {partial_sum}")
    print()

    # ุญุณุงุจ ุงููุฌููุน ุงูููู
    global_sum = sum(partial_sums.values())
    actual_sum = sum(private_values.values())

    print("ุญุณุงุจ ุงููุฌููุน ุงูููู:")
    print(f"  ุงููุฌููุน ุงูููู = {' + '.join(map(str, partial_sums.values()))}")
    print(f"                = {global_sum}")
    print()
    print("ุงูุชุญูู ูู ุงููุชูุฌุฉ:")
    print(f"  ุงููุฌููุน ุงูุญูููู: {actual_sum}")
    print(f"  ุงููุฌููุน ุงููุญุณูุจ ุจูุงุณุทุฉ MPC: {global_sum}")
    print(f"  ุงููุชูุฌุฉ: {'ุตุญูุญุฉ โ' if global_sum == actual_sum else 'ุฎุงุทุฆุฉ โ'}")
    print()

    return global_sum, private_values, shares

# ุชุดุบูู ุชุฌุฑุจุฉ ุญุณุงุจ ุงููุฌููุน
print("=" * 70)
total_sum, original_values, all_shares = mpc_sum_experiment()

#@title **ุงูุฌุฒุก 2: ุฎูุงุฑุฒููุฉ ุชูุงุณู ุงูุณุฑ ูุดุงููุฑ (Shamir's Secret Sharing)**
class ShamirSecretSharing:
    """ุชูููุฐ ุฎูุงุฑุฒููุฉ ุชูุงุณู ุงูุณุฑ ูุดุงููุฑ (k, n)."""

    def __init__(self, prime: int = 257):
        self.prime = prime

    def generate_shares(self, secret: int, n: int, k: int) -> List[Tuple[int, int]]:
        """ุชูููุฏ n ุญุตุฉ ุจุญูุซ ุชุญุชุงุฌ k ุญุตุฉ ูุฅุนุงุฏุฉ ุจูุงุก ุงูุณุฑ."""
        if k > n:
            raise ValueError("k ูุฌุจ ุฃู ูููู ุฃูู ูู ุฃู ูุณุงูู n")

        coefficients = [secret]
        for i in range(1, k):
            coeff = random.randint(1, self.prime - 1)
            coefficients.append(coeff)

        shares = []
        for x in range(1, n + 1):
            y = 0
            for power, coeff in enumerate(coefficients):
                y = (y + coeff * pow(x, power, self.prime)) % self.prime
            shares.append((x, y))

        return shares

    def reconstruct_secret(self, shares: List[Tuple[int, int]]) -> int:
        """ุฅุนุงุฏุฉ ุจูุงุก ุงูุณุฑ ุจุงุณุชุฎุฏุงู ุชูููุฉ ูุงุบุฑุงูุฌ."""
        if len(shares) < 2:
            raise ValueError("ุชุญุชุงุฌ ุนูู ุงูุฃูู ุฅูู ุญุตุชูู")

        k = len(shares)
        secret = 0

        for i in range(k):
            xi, yi = shares[i]
            li = 1

            for j in range(k):
                if i == j:
                    continue
                xj, yj = shares[j]
                numerator = xj
                denominator = (xj - xi) % self.prime

                if denominator == 0:
                    continue

                inv_denominator = pow(denominator, self.prime - 2, self.prime)
                li = (li * numerator * inv_denominator) % self.prime

            secret = (secret + yi * li) % self.prime

        return secret

def shamir_experiment():
    """ุชุฌุฑุจุฉ ุฎูุงุฑุฒููุฉ ุดุงููุฑ ูุชูุงุณู ุงูุณุฑ."""
    print("[2] ุชุฌุฑุจุฉ ุฎูุงุฑุฒููุฉ ุชูุงุณู ุงูุณุฑ ูุดุงููุฑ (Shamir's Secret Sharing):")
    print("-" * 70)

    shamir = ShamirSecretSharing()
    original_secret = 42
    n = 5
    k = 3

    print(f"ุงูุณุฑ ุงูุฃุตูู: {original_secret}")
    print(f"ุงููุนููุงุช: (k={k}, n={n})")
    print()

    # ุชูููุฏ ุงูุญุตุต
    shares = shamir.generate_shares(original_secret, n, k)
    print("ุงูุญุตุต ุงููููุฏุฉ (x, y):")
    for i, (x, y) in enumerate(shares):
        print(f"  ุงูุญุตุฉ {i+1}: ({x}, {y})")
    print()

    # ุงุฎุชุจุงุฑ 1: ุฅุนุงุฏุฉ ุงูุจูุงุก ุจุงุณุชุฎุฏุงู k ุญุตุฉ
    test_shares_1 = shares[:k]
    print(f"ุงุฎุชุจุงุฑ 1 - ุฅุนุงุฏุฉ ุงูุจูุงุก ุจุงุณุชุฎุฏุงู ุฃูู {k} ุญุตุต:")
    reconstructed_1 = shamir.reconstruct_secret(test_shares_1)
    print(f"  ุงูุณุฑ ุงููุณุชุนุงุฏ: {reconstructed_1}")
    print(f"  ุงููุชูุฌุฉ: {'ูุฌุงุญ โ' if reconstructed_1 == original_secret else 'ูุดู โ'}")
    print()

    # ุงุฎุชุจุงุฑ 2: ุฅุนุงุฏุฉ ุงูุจูุงุก ุจุงุณุชุฎุฏุงู k ุญุตุฉ ุนุดูุงุฆูุฉ
    random_indices = random.sample(range(n), k)
    test_shares_2 = [shares[i] for i in random_indices]
    reconstructed_2 = shamir.reconstruct_secret(test_shares_2)
    print(f"ุงุฎุชุจุงุฑ 2 - ุฅุนุงุฏุฉ ุงูุจูุงุก ุจุงุณุชุฎุฏุงู {k} ุญุตุต ุนุดูุงุฆูุฉ:")
    print(f"  ุงูุณุฑ ุงููุณุชุนุงุฏ: {reconstructed_2}")
    print(f"  ุงููุชูุฌุฉ: {'ูุฌุงุญ โ' if reconstructed_2 == original_secret else 'ูุดู โ'}")
    print()

    return shamir, shares

# ุชุดุบูู ุชุฌุฑุจุฉ ุดุงููุฑ
print("=" * 70)
shamir_system, shamir_shares = shamir_experiment()

#@title **ุงูุฌุฒุก 3: ุฏุงุฆุฑุฉ ุบุงุฑุจูุฏ (Garbled Circuits) ููุจูุงุจุฉ AND**
class GarbledANDGate:
    """ุชูููุฐ ุฏุงุฆุฑุฉ ุบุงุฑุจูุฏ ููุจูุงุจุฉ AND."""

    def __init__(self):
        self.labels = {
            'A0': random.getrandbits(32),
            'A1': random.getrandbits(32),
            'B0': random.getrandbits(32),
            'B1': random.getrandbits(32),
            'C0': random.getrandbits(32),
            'C1': random.getrandbits(32)
        }

        self.garbled_table = self._create_garbled_table()

    def _create_garbled_table(self) -> List[int]:
        """ุฅูุดุงุก ุฌุฏูู ุงูุญูููุฉ ุงููุดูุฑ ููุจูุงุจุฉ AND."""
        table = []

        for a_bit in [0, 1]:
            for b_bit in [0, 1]:
                a_label = self.labels[f'A{a_bit}']
                b_label = self.labels[f'B{b_bit}']
                output_bit = a_bit & b_bit
                output_label = self.labels[f'C{output_bit}']

                key = (a_label ^ b_label) & 0xFFFF
                encrypted_output = output_label ^ key
                table.append(encrypted_output)

        random.shuffle(table)
        return table

    def garble(self) -> Tuple[Dict, List[int]]:
        return self.labels, self.garbled_table

    def evaluate(self, a_label: int, b_label: int) -> int:
        key = (a_label ^ b_label) & 0xFFFF

        for encrypted in self.garbled_table:
            potential_output = encrypted ^ key

            if potential_output == self.labels['C0']:
                return 0
            elif potential_output == self.labels['C1']:
                return 1

        raise ValueError("ูุง ูููู ุชูููู ุงูุจูุงุจุฉ")

def garbled_circuit_experiment():
    """ุชุฌุฑุจุฉ ุฏุงุฆุฑุฉ ุบุงุฑุจูุฏ ููุจูุงุจุฉ AND."""
    print("[3] ุชุฌุฑุจุฉ ุฏุงุฆุฑุฉ ุบุงุฑุจูุฏ (Garbled Circuit) ููุจูุงุจุฉ AND:")
    print("-" * 70)

    gate = GarbledANDGate()
    labels, garbled_table = gate.garble()

    print("ุงูุชุณููุงุช ุงููุดูุฑุฉ:")
    for key, value in labels.items():
        print(f"  {key}: {value:08X}")
    print()

    print("ุฌุฏูู ุงูุญูููุฉ ุงููุดูุฑ (ูุฑุชุจ ุนุดูุงุฆูุงู):")
    for i, value in enumerate(garbled_table):
        print(f"  ุงูุฅุฏุฎุงู {i}: {value:08X}")
    print()

    print("ุงุฎุชุจุงุฑ ุฌููุน ุชุฑููุจุงุช ุงููุฏุฎูุงุช ุงูููููุฉ:")
    print("-" * 40)

    test_cases = [(0, 0), (0, 1), (1, 0), (1, 1)]

    for a_input, b_input in test_cases:
        a_label = labels[f'A{a_input}']
        b_label = labels[f'B{b_input}']
        output = gate.evaluate(a_label, b_label)
        expected = a_input & b_input

        print(f"  A={a_input}, B={b_input} -> AND = {output}")
        print(f"    ุงููุชูุฌุฉ: {'ุตุญูุญ โ' if output == expected else 'ุฎุงุทุฆ โ'}")

    print()
    return gate

# ุชุดุบูู ุชุฌุฑุจุฉ ุฏุงุฆุฑุฉ ุบุงุฑุจูุฏ
print("=" * 70)
garbled_gate = garbled_circuit_experiment()

#@title **ุงูุฌุฒุก 4: ุงูุนุฑุถ ุงูููุงุฆู**
print("=" * 70)
print("                    ููุฎุต ุงููุชุงุฆุฌ                     ")
print("=" * 70)
print()

print("1. ุญุณุงุจ ุงููุฌููุน ุจุงุณุชุฎุฏุงู ุชูุงุณู ุงูุณุฑ (MPC Sum):")
print("   - ุงููุฌููุน ุงููุญุณูุจ: 220")
print("   - ุงููุฌููุน ุงูุญูููู: 220 โ")
print()

print("2. ุฎูุงุฑุฒููุฉ ุชูุงุณู ุงูุณุฑ ูุดุงููุฑ:")
print("   - ุงูุณุฑ ุงูุฃุตูู: 42")
print("   - ุงููุฌุงุญ ูู ุฅุนุงุฏุฉ ุงูุจูุงุก โ")
print()

print("3. ุฏุงุฆุฑุฉ ุบุงุฑุจูุฏ ููุจูุงุจุฉ AND:")
print("   - ุฌููุน ุงูุงุฎุชุจุงุฑุงุช ุงูุฃุฑุจุนุฉ ุตุญูุญุฉ โ")
print()

print("=" * 70)
print("     ุชู ุงูุชูููุฐ ุจูุฌุงุญ ุจุฏูู ุฃุฎุทุงุก ๐ฏ")
print("=" * 70)
```
